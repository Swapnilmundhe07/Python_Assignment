## Question 5: Sum of Elements Greater Than Current Element

You are given a list of integers. Create a list where each element is replaced by the sum of itself and all greater elements in the list.

Example:

Input:

[2,4,6,10,1]

Output:

[22,20,16,10,23]

```python
L=[2,4,6,10,1]
N=[]
for i in L:
    total=0
    for j in L:
        if j>=i:
            total=total+j
    N.append(total)
print(N)
```

## Question 11: Count Unique Vowels Using Sets

Write a program to count unique vowels in a string using sets. Lowercase and uppercase vowels should be treated differently.

Example:

Input:

"hands-on data science mentorship progrAm with live classes at affordable fee only on CampusX"

Output:

No of unique vowels = 6

```python
def count_vowels(text):
    vowels=set('aeiouAEIOU')
    unique_vowels=set()
    for i in text:
     if i in vowels:
        unique_vowels.add(i)
    return unique_vowels,len(unique_vowels)

text="hands-on data science mentorship progrAm with live classes at affordable fee only on CampusX"
a=count_vowels(text)
print(a)
```

## Question 13: Convert List of Tuples into Dictionary

Convert a list of tuples into dictionary format.

Example:

Input:

[("akash",10),("gaurav",12),("anand",14)]

Output:

{'akash':[10],'gaurav':[12],'anand':[14]}

```python
def convert(tup_dic):
    result={}
    for k,v in tup_dic:
        result[k]=[v]
    return result
a=[("akash",10),("gaurav",12),("anand",14)]
b=convert(a)
print(b)
```

## Question 14: Unique Elements from List

Write a Python function that takes a list and returns a new list with unique elements.

Example:

Input:

[1,2,3,3,3,4,5]

Output:

[1,2,3,4,5]

```python

def unique(lst):
    new = []
    for i in lst:
        if i not in new:
            new.append(i)
    return new
lst = [1, 2, 3, 3, 3, 4, 5]
result = unique(lst)
print(result)
```

## Question 16: Check Perfect Number

Write a Python function to check whether a number is perfect or not.

Example:

6 is a perfect number because:

1 + 2 + 3 = 6

```python
def perfect(n):
    total=0
    for i in range(1,n):
        if n%i==0:
            total=total+i
    if total==n:
        return True
    else:
        return False
    
a=perfect(6)
print(a)
```

## Question 17: Concatenate Multiple Dictionaries

Write a Python function to concatenate any number of dictionaries to create a new one.

Example:

dic1={1:10,2:20}  
dic2={3:30,4:40}  
dic3={5:50,6:60}

Output:

{1:10,2:20,3:30,4:40,5:50,6:60}

```python
def concat(*args):
    new={}
    for i in args:
        new.update(i)
    return new
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}
a=concat(dic1,dic2,dic3)
print(a)
```
