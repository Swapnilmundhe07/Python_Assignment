```python
import math

# Base class
class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass


# Circle class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

    def perimeter(self):
        return 2 * math.pi * self.radius


# Square class
class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side

    def perimeter(self):
        return 4 * self.side


# Triangle class
class Triangle(Shape):
    def __init__(self, a, b, c, height, base):
        self.a = a
        self.b = b
        self.c = c
        self.height = height
        self.base = base

    def area(self):
        return 0.5 * self.base * self.height

    def perimeter(self):
        return self.a + self.b + self.c


# Creating objects
c = Circle(5)
s = Square(4)
t = Triangle(3, 4, 5, 4, 3)

# Output
print("Circle Area:", c.area())
print("Circle Perimeter:", c.perimeter())

print("Square Area:", s.area())
print("Square Perimeter:", s.perimeter())

print("Triangle Area:", t.area())
print("Triangle Perimeter:", t.perimeter())
```

8. Write a Python program to create a class representing a shopping cart. Include methods for adding and removing items, and calculating the total price. 


```python
class ShoppingCart:
    def __init__(self):
        self.cart = {}   # dictionary to store item and price

    def add_item(self, item, price):
        self.cart[item] = price #self.cart[item] is this key or value
        #self.price=price
        print(item, "added to cart")

    def remove_item(self, item):
        if item in self.cart:
            del self.cart[item] #self.cart[item] is this key or value , if key then what is being deleted
            print(item, "removed from cart")
        else:
            print(item, "not found in cart")

    def total_price(self):
        total = sum(self.cart.values())
        print("Total price =", total)


# Creating object
c = ShoppingCart()

# Using methods
c.add_item("Shoes", 2000)
c.add_item("Bag", 1500)
c.add_item("Watch", 3000)

c.remove_item("Bag")

c.total_price()
```

```python
class Bank:
    def __init__(self,name,acc_no,balance=0):
        self.name=name
        self.acc_no=acc_no
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
        print(f'customer has deposited:{ self.balance}')
    def withdraw(self,amount):
        if amount > self.balance:
            print('Insufficient amount')
        else:
            self.balance-=amount
            print(f'amount after withdraw:{self.balance}')
    def annual_fee(self):
        fee=0.05*self.balance
        self.balance-=fee
        print(f'annual fees paid by customer is ={0.05*self.balance}')
        print( f'remaining balance after annual fees is={self.balance}')
    def display(self):
        print(f'Name of customer:{self.name}\n Account number of customer {self.acc_no}')
        print(f'balancde of customer:{self.balance}')
c=Bank('Admin Patil',123)
#c.display()
c.deposit(50000)
c.withdraw(25000)
c.annual_fee()
c.display()
```
